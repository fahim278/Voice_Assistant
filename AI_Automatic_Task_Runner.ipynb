# üé§ Step 2: Import Libraries

```python
import speech_recognition as sr
from gtts import gTTS
from pydub import AudioSegment
from google.colab import files
import IPython.display as ipd
import os
```

# üß† Step 3: Helper ‚Äî Speak Functio

```python
def speak(text):
    print("ü§ñ AI Assistant:", text)
    tts = gTTS(text=text, lang='en')
    tts.save("response.mp3")
    return ipd.Audio("response.mp3", autoplay=True)
```

# üéôÔ∏è Step 4: Upload an Audio File
1.   You‚Äôll record yourself saying something like
      ‚ÄúOpen browser‚Äù
      ‚ÄúPlay music‚Äù
      ‚ÄúTake screenshot‚Äù
      ‚ÄúHello assistant‚Äù
      and upload it (must be .mp3 or .wav).

```python
uploaded = files.upload()
audio_path = list(uploaded.keys())[0]
print(f"üéß Uploaded audio: {audio_path}")

# Convert to .wav if needed
if audio_path.endswith('.mp3'):
    sound = AudioSegment.from_mp3(audio_path)
    audio_path = "converted.wav"
    sound.export(audio_path, format="wav")
```

# üß© Step 5: Recognize Speech and Process Commands

```python
def process_command(command):
    command = command.lower()
    print(f"üéØ Command Detected: {command}")

    if "open browser" in command:
        return "Opening browser... (simulated)"
    elif "play music" in command:
        return "Playing your favorite music... (simulated)"
    elif "take screenshot" in command:
        return "Taking a screenshot... (simulated)"
    elif "open notepad" in command:
        return "Opening notepad... (simulated)"
    elif "hello" in command:
        return "Hello Fahim! How can I assist you today?"
    else:
        return "Sorry, I didn‚Äôt understand that command."

r = sr.Recognizer()
with sr.AudioFile(audio_path) as source:
    speak("Listening to your voice command...")
    audio = r.record(source)

try:
    command = r.recognize_google(audio)
    action_result = process_command(command)
    audio_output = speak(action_result)
    display(audio_output)
except sr.UnknownValueError:
    speak("Sorry, I could not understand the audio.")
except sr.RequestError:
    speak("Speech recognition service failed.")
```
